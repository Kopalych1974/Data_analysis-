```python
import pandas as pd

DATA_SAMPLE_CSV_URL = 'Data Sample (AIC) для итоговой работы.txt'

df = pd.read_csv(DATA_SAMPLE_CSV_URL, sep=";", header=0, index_col=False)
df["UnitPrice"] = df['UnitPrice'].astype('float')
df["Country"] = df["Country"].astype('category')
df
```

Добавим столбец с итоговой суммой по каждой покупке

```python
TotalAmount = df['Quantity'] * df['UnitPrice']
df.insert(loc=7,column='TotalAmount',value=TotalAmount)
df
```

Сгруппируем данные для удобства анализа.

```python
df = (df[['InvoiceDate', 'Country', 'CustomerID', 'InvoiceNo', "StockCode", 'Description', 'Quantity', 'UnitPrice', 'TotalAmount']])
df
```

Начнем с простого — проанализируем суммарную статистику предоставленного датасета. 

```python
df.info()
```

Из суммарной статистики видно, что CustomerID заполнено не везде. Уберем такие данные из анализа.

```python
cleaned_retail = df.loc[pd.isnull(df.CustomerID) == False]


print(cleaned_retail.info())
```

```python
#show the min and max quantites has sold. 
final_df = df[(df['Quantity'] > 0) & (df['TotalAmount'] > 0)]
final_df
```

```python
print("Итоговое значение строк для анализа: ", final_df.shape[0])
print("Итоговое значение не включенных в анализ данных: {} - which is {:.2f}% of our total data".format(final_df.isnull().sum().sum(), (final_df.isnull().sum().sum()*100)/final_df.shape[0]))
print("Итоговое значение уникальных стран: ", final_df.Country.nunique())
print("Итоговое значение уникальных товаров: ", final_df.Description.nunique())
```

```python
# Перечень наименее популярных товаров
final_df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).tail(700)
```

```python
# Перечень наиболее популярных товаров
final_df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(50)
```
